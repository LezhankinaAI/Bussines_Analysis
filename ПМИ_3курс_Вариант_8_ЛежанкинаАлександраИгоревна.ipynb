```python
from sklearn.model_selection import train_test_split


# Убираю колонку UID, так как она не несет в себе полезную информацию о пользователе
# Выделяю также валидационную выборку для поиска по ней значений WoE, чтобы не переобучаться
X_train, X_test, y_train, y_test = train_test_split(accept.drop(columns=['target', 'UID']), accept['target'], 
                                                    test_size=0.2, shuffle=True, random_state=66)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.6, shuffle=True, random_state=10)
```

```python
# начинаем подсчёт WoE по валидационной выборке, а также соответствующая инициализация для accept и reject

inf_values = {}

for num in numeric:
    df = pd.DataFrame(index=X_valid.index, columns=['bin', 'value_counts', '1', 'f(x|y=1)', '0', 'f(x|y=0)', 'WoE'])
    X_train['bin'] = -1
    X_test['bin'] = -1
    X_valid['bin'] = -1
    reject['bin'] = -1
    all_0 = y_valid.value_counts()[0]
    all_1 = y_valid.value_counts()[1]
    minim = np.min([reject[num].min(), X_train[num].min(), X_valid[num].min(), X_test[num].min()])
    maxim = np.max([reject[num].max(), X_train[num].max(), X_valid[num].max(), X_test[num].max()]) + 1
    diff = (maxim - minim) / 20
    # делю изначально выборку на 20 интервалов, затем объединяю сходие WoE    
    for i in range(20):
        val = minim + diff * i
        val_next = minim + diff * (i + 1)
        # определяем бины для каждой выборки, чтобы в дальнейшем подставить туда подходящее WoE        
        df.loc[(X_valid[num] >= val) & (X_valid[num] < val_next), 'bin'] = i
        X_valid.loc[(X_valid[num] >= val) & (X_valid[num] < val_next), 'bin'] = i
        X_train.loc[(X_train[num] >= val) & (X_train[num] < val_next), 'bin'] = i
        X_test.loc[(X_test[num] >= val) & (X_test[num] < val_next), 'bin'] = i
        reject.loc[(reject[num] >= val) & (reject[num] < val_next), 'bin'] = i
        df.loc[df.bin == i, 'value_counts'] = (df.bin == i).sum()
        # далее считаем необходимые значения, если объект в данном интервале присутвтует        
        if (df.bin == i).sum() != 0:
            df.loc[df.bin == i, '1'] = df.loc[(y_valid == 1) & (df.bin == i), 'bin'].count()
            df.loc[df.bin == i, '0'] = df.loc[(y_valid == 0) & (df.bin == i), 'bin'].count()
            if all(df.loc[df.bin == i, 'value_counts']) == 0:
                df.loc[df.bin == i, 'f(x|y=1)'] = 0.0001
                df.loc[df.bin == i, 'f(x|y=0)'] = 0
            else:
                df.loc[df.bin == i, 'f(x|y=1)'] = df.loc[df.bin == i, '1'] / all_1 + 0.0001 # добавляем небольшую константу,
                df.loc[df.bin == i, 'f(x|y=0)'] = df.loc[df.bin == i, '0'] / all_0 + 0.0001 # если 1 или 0 в бине нет, 
    df.WoE = np.log((df['f(x|y=0)'] / df['f(x|y=1)']).astype(np.float64))                   # чтобы не делить далее на 0
    IV = 0
    # начинаем процесс объединения бинов с маленькой разницей WoE. Я выбрала такой разницей 0.05
    df.sort_values(by='WoE', inplace=True)
    bins = df.bin.unique()
    size = df.bin.nunique() - 1
    i = 0
    ints = {i : [i] for i in df.bin.unique()}
    while i < size and size > 5:
        bins = df.bin.unique()
        if (abs(df.loc[df.bin == bins[i], 'WoE'].iloc[0] - df.loc[df.bin == bins[i + 1], 'WoE'].iloc[0])) <= 0.05:
            ints[bins[i]].append(bins[i + 1])
            df.loc[df.bin.isin([bins[i], bins[i + 1]]), ['value_counts', '1', '0']] = \
                df.loc[df.bin.isin([bins[i], bins[i + 1]])].drop_duplicates().sum()[['value_counts', '1', '0']].to_list()
            df.loc[df.bin == bins[i + 1], 'bin'] = bins[i]
            df.loc[df.bin == bins[i], 'f(x|y=1)'] = df.loc[df.bin == bins[i], '1'] / all_1 + 0.0001
            df.loc[df.bin == bins[i], 'f(x|y=0)'] = df.loc[df.bin == bins[i], '0'] / all_0 + 0.0001
            df.loc[df.bin == bins[i], 'WoE'] = np.log((df.loc[df.bin == bins[i],'f(x|y=0)'] / \
                                                       df.loc[df.bin == bins[i],'f(x|y=1)']).astype(np.float64))
            i -= 1
            size -= 1
        i += 1
    name = 'Woe_' + num
    # инициализируем WoE у accept и reject
    for bn in df.bin.unique():        
        X_train.loc[X_train['bin'].isin([*ints[bn]]), name] = df.loc[df.bin == bn, 'WoE'].iloc[0]
        X_test.loc[X_test['bin'].isin([*ints[bn]]), name] = df.loc[df.bin == bn, 'WoE'].iloc[0]
        X_valid.loc[X_valid['bin'].isin([*ints[bn]]), name] = df.loc[df.bin == bn, 'WoE'].iloc[0]
        reject.loc[reject['bin'].isin([*ints[bn]]), name] = df.loc[df.bin == bn, 'WoE'].iloc[0]
    # считаем IV для accept целиком (так как до этого сплитили на трейн, тест, вэлид, чтобы искать WoE по отложенной выборке,
    # сейчас приходится обратно конкатенировать выборки)
    accept_IV = pd.concat([X_train, X_valid, X_test])
    y = pd.concat([y_train, y_valid, y_test])
    for bn in df.bin.unique():
        accept_IV.loc[accept_IV.bin == bn, 'f(x|y=1)'] = accept_IV.loc[(y == 1) & (accept_IV.bin == bn), 'bin'].count() \
                / y.value_counts()[1] + 0.0001
        accept_IV.loc[accept_IV.bin == bn, 'f(x|y=0)'] = accept_IV.loc[(y == 0) & (accept_IV.bin == bn), 'bin'].count() \
                / y.value_counts()[0] + 0.0001        
        IV += (accept_IV.loc[accept_IV.bin == bn, 'f(x|y=0)'].iloc[0] - accept_IV.loc[accept_IV.bin == bn, 'f(x|y=1)'].iloc[0]) * \
                accept_IV.loc[accept_IV.bin == bn, name].iloc[0]
    # костыль, так как если заметить, в выборке reject максимальное значение возраста меньше минимального из accept,
    # то есть посчитать для данного интервала значение WoE по отложенной выборке невозможно, ведь подобных значений
    # там не встречается. Поэтому я инициализирую весь признак age в reject WoE для минимльного бина из accept
    if num == 'age':
        reject[name] = df.loc[df['bin'] == df['bin'].min(), 'WoE'].iloc[0]
    if IV == -np.inf:
        IV = 0
    inf_values[num] = IV
    print(num, 'IV = %.3f' % IV)
```

```python
# алгоритм действий для категориальных переменных совпадает с интервальными за тем исключением, что мы не определяем интервалы
# вручную. Значениями бинов являются соответствующие уникальные категории у признака. 
for cat in categorical:
    df = pd.DataFrame(index=X_valid.index, columns=['bin', 'value_counts', '1', 'f(x|y=1)', '0', 'f(x|y=0)', 'WoE'])
    X_train['bin'] = ''
    X_test['bin'] = ''
    X_valid['bin'] = ''
    all_0 = y_valid.value_counts()[0]
    all_1 = y_valid.value_counts()[1]
    for c in (X_valid[cat].unique()):
        df.loc[(X_valid[cat] == c), 'bin'] = c
        X_valid.loc[(X_valid[cat] == c), 'bin'] = c
        X_train.loc[(X_train[cat] == c), 'bin'] = c
        X_test.loc[(X_test[cat] == c), 'bin'] = c
        reject.loc[(reject[cat] == c), 'bin'] = c
        df.loc[df.bin == c, 'value_counts'] = (df.bin == c).sum()
        if (df.bin == c).sum() != 0:
            df.loc[df.bin == c, '1'] = df.loc[(y_valid == 1) & (df.bin == c), 'bin'].count()
            df.loc[df.bin == c, '0'] = df.loc[(y_valid == 0) & (df.bin == c), 'bin'].count()
            if all(df.loc[df.bin == c, 'value_counts']) == 0:
                df.loc[df.bin == c, 'f(x|y=1)'] = 0.0001
                df.loc[df.bin == c, 'f(x|y=0)'] = 0
            else:
                df.loc[df.bin == c, 'f(x|y=1)'] = df.loc[df.bin == c, '1'] / all_1 + 0.0001
                df.loc[df.bin == c, 'f(x|y=0)'] = df.loc[df.bin == c, '0'] / all_0 + 0.0001
    df.WoE = np.log((df['f(x|y=0)'] / df['f(x|y=1)']).astype(np.float64))
    IV = 0
    df.sort_values(by='WoE', inplace=True)
    bins = df.bin.unique()
    size = df.bin.nunique() - 1
    i = 0
    ints = {i : [i] for i in df.bin.unique()}
    while i < size and size > 5:
        bins = df.bin.unique()
        if (abs(df.loc[df.bin == bins[i], 'WoE'].iloc[0] - df.loc[df.bin == bins[i + 1], 'WoE'].iloc[0])) <= 0.05:
            ints[bins[i]].append(bins[i + 1])
            df.loc[df.bin.isin([bins[i], bins[i + 1]]), ['value_counts', '1', '0']] = \
                df.loc[df.bin.isin([bins[i], bins[i + 1]])].drop_duplicates().sum()[['value_counts', '1', '0']].to_list()
            df.loc[df.bin == bins[i + 1], 'bin'] = bins[i]
            df.loc[df.bin == bins[i], 'f(x|y=1)'] = df.loc[df.bin == bins[i], '1'] / all_1 + 0.0001
            df.loc[df.bin == bins[i], 'f(x|y=0)'] = df.loc[df.bin == bins[i], '0'] / all_0 + 0.0001
            df.loc[df.bin == bins[i], 'WoE'] = np.log((df.loc[df.bin == bins[i],'f(x|y=0)'] / \
                                                       df.loc[df.bin == bins[i],'f(x|y=1)']).astype(np.float64))
            i -= 1
            size -= 1
        i += 1
    name = 'Woe_' + cat
    for bn in df.bin.unique():        
        X_train.loc[X_train['bin'].isin([*ints[bn]]), name] = df.loc[df.bin == bn, 'WoE'].iloc[0]
        X_test.loc[X_test['bin'].isin([*ints[bn]]), name] = df.loc[df.bin == bn, 'WoE'].iloc[0]
        X_valid.loc[X_valid['bin'].isin([*ints[bn]]), name] = df.loc[df.bin == bn, 'WoE'].iloc[0]
        reject.loc[reject['bin'].isin([*ints[bn]]), name] = df.loc[df.bin == bn, 'WoE'].iloc[0]
    accept_IV = pd.concat([X_train, X_valid, X_test])
    y = pd.concat([y_train, y_valid, y_test])
    for bn in df.bin.unique():
        accept_IV.loc[accept_IV.bin == bn, 'f(x|y=1)'] = accept_IV.loc[(y == 1) & (accept_IV.bin == bn), 'bin'].count() \
                / y.value_counts()[1] + 0.0001
        accept_IV.loc[accept_IV.bin == bn, 'f(x|y=0)'] = accept_IV.loc[(y == 0) & (accept_IV.bin == bn), 'bin'].count() \
                / y.value_counts()[0] + 0.0001        
        IV += (accept_IV.loc[accept_IV.bin == bn, 'f(x|y=0)'].iloc[0] - accept_IV.loc[accept_IV.bin == bn, 'f(x|y=1)'].iloc[0]) * \
                accept_IV.loc[accept_IV.bin == bn, name].iloc[0]
    if IV == -np.inf:
        IV = 0
    inf_values[cat] = IV
    print(cat, 'IV = %.3f' % IV)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score, f1_score



features = ['Woe_income',
       'Woe_DEPENDANT_NUMBER', 'Woe_COUNT_ACTIVE_CREDIT_NO_CC',
       'Woe_CNT_MNTH_FROM_LAST_PMNT', 'Woe_age', 'Woe_experience',
       'Woe_inquiry_14_day', 'Woe_inquiry_21_day',
       'Woe_count_mnth_act_passport', 'Woe_INCOME_TYPE', 'Woe_EDUCATION_',
       'Woe_CUSTOMER_CATEGORY_', 'Woe_IS_EMPL']

parameters = {'penalty':['l2'], 'C':np.logspace(-3, 2, 5)}

model = GridSearchCV(LogisticRegression(), parameters)
model.fit(X_train[features], y_train)

best_c1 = model.best_params_['C']
print("Best c = %.5f" % best_c1)

classifier1 = LogisticRegression(C=best_c1)
classifier1.fit(X_train[features], y_train)
print('auc roc:', roc_auc_score(y_test, classifier1.decision_function(X_test[features])))
print('Gini score:', 2 * roc_auc_score(y_test, classifier1.decision_function(X_test[features])) - 1)
score = f1_score(y_test, (classifier1.predict_proba(X_test[features])[:, 0] < 0.9775).astype(int))
print('f1 score:', score)
```

```python
# здесь считаю IV для reject + accept

columns = ['Woe_income',
       'Woe_DEPENDANT_NUMBER', 'Woe_COUNT_ACTIVE_CREDIT_NO_CC',
       'Woe_CNT_MNTH_FROM_LAST_PMNT', 'Woe_age', 'Woe_experience',
       'Woe_inquiry_14_day', 'Woe_inquiry_21_day',
       'Woe_count_mnth_act_passport', 'Woe_INCOME_TYPE', 'Woe_EDUCATION_',
       'Woe_CUSTOMER_CATEGORY_', 'Woe_IS_EMPL']

ar_IV = pd.concat([X_train, X_valid, X_test, reject.drop(columns='target')])
ar_y = pd.concat([y_train, y_valid, y_test, reject.target])
for feat in columns:
    arIV = 0
    for w in ar_IV[feat].unique():
        ar_IV.loc[ar_IV[feat] == w, 'f(x|y=1)'] = ar_IV.loc[(ar_y == 1) & (ar_IV[feat] == w), feat].count() \
                / ar_y.value_counts()[1] + 0.0001
        ar_IV.loc[ar_IV[feat] == w, 'f(x|y=0)'] = ar_IV.loc[(ar_y == 0) & (ar_IV[feat] == w), feat].count() \
                / ar_y.value_counts()[0] + 0.0001        
        arIV += abs((ar_IV.loc[ar_IV[feat] == w, 'f(x|y=0)'].iloc[0] - ar_IV.loc[ar_IV[feat] == w, 'f(x|y=1)'].iloc[0]) * \
                ar_IV.loc[ar_IV[feat] == w, feat].iloc[0])
    print(feat, 'IV = %.3f' % arIV)
```
